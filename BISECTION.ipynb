```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Persamaan Awal
def f(x):
  y= x**3 - 3*x**2 - 3*x + 1
  return y
#Pembagian persamaan awal menjadi 2 bagian
def g(x):
 y1 = x**3
 return y1
def h(x):
 y2 = 3*x**2 + 3*x - 1
 return y2
#Tebakkan nilai awal
a = -5
b = 5
#Rumus Bisection
def bisection(a, b):
 if f(a)*f(b)>0:
  print("Tidak Ada Akar Akar")
  return
 c = a
 while ((b-a)>=0.01):
  c = (a +b)/2
  if f(c)==0:
    exit()
  if f(c)*f(a)<0:
     b = c
  else:
    a = c
 print("root of i = ", c)
 plt.plot(c , f(c), "ro")
bisection(a, b)
x=np.linspace(-3,3,20)
#Grafik Ploting
plt.plot (x, g(x), x, h(x), x, f(x))
plt.grid()
#Penamaan Label
plt.xlabel("x", fontsize = 15)
plt.ylabel("y", fontsize = 15, rotation = 0)
plt.text(-2,7, "Perpotongan antara $y_1$ dan $y_2$")
plt.text(-2.5,-40, "$y=x^3 - 3x^2 - 3x + 1$")
plt.text(2,30, "$y_1= x^3$")
plt.text(1.2,-2, "$y_2= 3x^2 + 3x - 1$")
```
